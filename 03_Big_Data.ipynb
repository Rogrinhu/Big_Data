```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Data': ['Jan 2023', 'Fev 2023', 'Mar 2023', 'Abr 2023', 'Mai 2023'],
    'Exercícios Militares': [2, 3, 4, 3, 5],
    'Medidas de Controle': [1, 1, 2, 2, 3]}
```

```python
df = pd.DataFrame(data)
```

```python
exercicios_militares = pd.Series(df['Exercícios Militares'].values, index=df['Data'])
medidas_controle = pd.Series(df['Medidas de Controle'].values, index=df['Data'])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(exercicios_militares, label='Exercícios Militares', marker='o')
plt.plot(medidas_controle, label='Medidas de Controle', marker='o')
plt.title('Tensão Nuclear - Exercícios Militares vs. Medidas de Controle')
plt.xlabel('Data')
plt.ylabel('Quantidade')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
